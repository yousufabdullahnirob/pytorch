```python
import torch

a = torch.tensor(2.0)
b = torch.tensor(3.0, requires_grad=True)

y_pred = a * b
y_true = torch.tensor(10.0)

print(y_pred)

loss = (y_pred - y_true) ** 2
print(loss)

loss.backward()


print(b.grad)
print (a.grad)#a cannot learn
```

```python
import torch

# constants / parameters
a = torch.tensor(2.0)                      # constant
b = torch.tensor(-1.0, requires_grad=True) # learnable parameter

# forward pass
y_pred = a * b
y_true = torch.tensor(-22.0)

print("y_pred:", y_pred.item())

# loss
loss = (y_pred - y_true) ** 2
print("loss:", loss.item())

# backward pass
loss.backward()

# gradient
print("b.grad:", b.grad)   # correct
print("a.grad:", a.grad)   # None (cannot learn)

# parameter update (Gradient Descent)
learning_rate = 0.1

with torch.no_grad():
    b -= learning_rate * b.grad

print("updated b:", b.item())

# reset gradient for next iteration
b.grad.zero_()
print("b.grad after zero:", b.grad)

```
