```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1Ô∏è‚É£ Define the model
class RiskModel(nn.Module):
    def __init__(self):
        super().__init__()
        # 3 input features: height, weight, age
        # 1 output: risk_score
        self.linear = nn.Linear(3, 1)

    def forward(self, x):
        return self.linear(x)


# 2Ô∏è‚É£ Create model object
model = RiskModel()

# 3Ô∏è‚É£ Dummy input data (batch_size = 2)
# shape: (batch_size, input_features) = (2, 3)
x = torch.tensor([
    [170.0, 65.0, 30.0],   # person 1
    [160.0, 70.0, 45.0]    # person 2
])

# 4Ô∏è‚É£ Target risk scores
# shape: (batch_size, output_features) = (2, 1)
y_true = torch.tensor([
    [0.8],
    [0.6]
])

# 5Ô∏è‚É£ Forward pass
y_pred = model(x)

# 6Ô∏è‚É£ Loss function
criterion = nn.MSELoss()

# 7Ô∏è‚É£ Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.0001)

# 8Ô∏è‚É£ Compute loss
loss = criterion(y_pred, y_true)

# 9Ô∏è‚É£ Backpropagation
optimizer.zero_grad()
loss.backward()
optimizer.step()

# üîü Print results
print("Predicted risk scores:\n", y_pred)
print("Loss:", loss.item())
print("Weights shape:", model.linear.weight.shape)
print("Bias shape:", model.linear.bias.shape)
```
