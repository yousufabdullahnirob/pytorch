```python
import torch

x = torch.tensor([1,2,3])
y_true = torch.tensor([4,5,6])#correct values

print(x + y_true)



w = torch.tensor(0.0,requires_grad=True)

lr = 0.01
epochs = 10

for epoch in range(epochs):
    #step1 prediction
    y_pred = w * x

    #step2 loss
    loss = (y_pred - y_true).pow(2).mean()
    
    loss.backward()

    with torch.no_grad():
        w.data = w.data - lr * w.grad.data
        print(f"Epoch {epoch+1}:w = {w.item():4f}, loss = {loss.item():4f}")
    
    w.grad.zero_()
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```
