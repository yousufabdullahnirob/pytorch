```python
import torch
import torch.nn as nn

class MyModel(nn.Module):  
    # inherit from nn.Module
    def __init__(self):
        super().__init__()          # initialize nn.Module
        self.w = nn.Parameter(torch.tensor(0.0))  # learnable parameter

    def forward(self, x): #self means own model 
        return self.w * x #what need to provide

```
