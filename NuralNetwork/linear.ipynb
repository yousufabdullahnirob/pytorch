```python
import torch
import torch.nn as nn

class RiskModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(3, 1)  # 3 inputs â†’ 1 output

    def forward(self, x):
        return self.linear(x)

```

```python
model = RiskModel()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# height, weight, age
x = torch.tensor([170.0, 65.0, 30.0])

risk_score = model(x)
print(risk_score)
```
