```python
import torch
#step1 

x = torch.tensor([1.,2.,3.])

#step2
y = torch.tensor([2.,4.,6.])

#step3 worng guess

w = torch.tensor(0.0)
b = torch.tensor(0.0)

#step4

y_pred = w*x+b

loss = ((y_pred-y)**2).mean()#mean squared error

print("predict value:",y_pred)
print("loss value:",loss)
print ("actual value:",y)
```
